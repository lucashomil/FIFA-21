## Building Our Dream Team

Using Genetic Algorithm to build our dream, the Algorithm will select a team based on some features such as total amount of value and average age of players.

```python
import numpy as np

import pandas as pd

import random

import matplotlib.pyplot as plt
```

```python
# Loading data set - already cleaned

df = pd.read_excel('./Dataset/Dataset-FIFA-21-cleaned.xlsx')

df.head()
```

#### Some columns will be only used. And, we will only select players who have good values. Thus, we are going to trim down the DataFrame.

```python
df = df.loc[:, ['Name', 'Age', 'Overall', 'Value (€)', 'Position']]

df = df.loc[df['Value (€)'] > 10000]

df.reset_index(drop = True, inplace = True)

df.head()
```

#### In order to simplify task, we will clariy next positions: Center Forward, Left Forward, Right Forward, Center Half, Left Half, Right Half, Center Defender, Left Defender, Right Defender and Goalkeeper. We will select two teams (main and second). Thus, there will be 22 players.

```python
replace_dict = {
    'LB': 'Left Defender','LCB': 'Center Defender', 'SUB': 'Center Half', 'RW': 'Right Half',
    'RES': 'Center Half', 'CAM': 'Center Half', 'RCM': 'Center Half', 'LW': 'Left Half',
    'RCB': 'Center Defender', 'RF': 'Right Forward', 'LCM': 'Center Half', 'LS': 'Left Forward',
    'RB': 'Right Defender', 'RM': 'Right Half', 'ST': 'Center Forward', 'LWB': 'Left Defender',
    'LM': 'Left Half', 'RWB': 'Right Defender', 'RS': 'Right Forward', 'GK': 'Goalkeeper',
    'CDM': 'Center Half', 'CF': 'Center Forward', 'CB': 'Center Defender', 'LF': 'Left Forward', 
    'CM': 'Center Half', 'RDM': 'Center Half', 'LAM': 'Left Half', 'LDM': 'Center Half', 'RAM': 'Right Half'}

df['Position'].replace(replace_dict, inplace = True)
```

#### There are many formations in soccer. In theory, a formation describes how the players in a team are positioned on the field. We will use the most common formation for our team (4-3-3: 4 Defenders, 3 Halfs and 3 Forwards).

We will have a class to describe some basic information of a player.

```python
class Player:
    
    def __init__(self, name, age, overall, value, position):
        
        self.name = name
        
        self.age = age
        
        self.overall = overall
        
        self.value = value
        
        self.position = position
```

#### We will have a function in order to randomly select 2 unique players for every position. Only for position Central Defender we will select 4 players, because we need 2 of them on the field at the same time.

```python
def generate_player(player_df, genes, use_best = False):
    
    gen_df = player_df.sort_values(['Overall'], ascending = False)
    
    if use_best:
        
        player = 0
        
    else:
        
        player = random.randint(0, len(gen_df)-1)
        
    player_item = Player(
        gen_df.iloc[player]['Name'], 
        gen_df.iloc[player]['Age'],
        gen_df.iloc[player]['Overall'],
        gen_df.iloc[player]['Value (€)'],
        gen_df.iloc[player]['Position']
    )
    
    return player_item
```

```python
def create_individual(football_df):
    
    genes = []
    
    positions = [ 'Goalkeeper', 'Left Defender', 'Right Defender', 'Center Defender', 'Left Half', 'Right Half', 
                'Center Half', 'Left Forward', 'Right Forward', 'Center Forward'] 
    
    for pos in positions:
        
        player_df = football_df.loc[football_df['Position'] == pos]
        
        for i in range(2):
            
            player_item = generate_player(player_df, genes)
            
            genes.append(player_item)
            
        if pos == 'Center Defender':
            
            for i in range(2):
                
                player_item = generate_player(player_df, genes)
                
                genes.append(player_item)

    return genes
```

#### Our fitness function is a total power of a team (sum of Overall) minus amount of money that we spent over our limits and minus number of years * 10 over our limit (average age).

```python
def fitness(max_money, avg_age, individual):
    
    score = cost = age = 0
    
    for player in individual:
        
        score += player.overall
        
        cost += player.value
        
        age += player.age

    score += min(0, max_money-cost)
    
    age_diff = (avg_age - age / 22) * 10
    
    score += min(0, age_diff)
    
    if len(list(set([it.name for it in individual]))) != 22:
        
        score -= 1000000
        
    return score
```

#### To mutate individual we just select new player for current position if random value is less than mutation rate.

```python
def mutate(individual, mutation_rate, mutation_best_rate, footbal_df):
    
    new = []
    
    for gene in individual:
        
        player_df = footbal_df.loc[footbal_df['Position'] == gene.position]
        
        if mutation_best_rate > random.random():
            
            player_item = generate_player(player_df, individual, use_best=True)
            
            new.append(player_item)
            
            continue
            
        if mutation_rate > random.random():
            
            player_item = generate_player(player_df, individual)
            
            new.append(player_item)
            
        else:
            
            player_item = generate_player(player_df, individual)
            
            new.append(player_item)
            
    return new
```

#### Let's define the features

```python
MAX_MONEY = 1200000000

AVG_AGE = 25

EPOCHS = 800

CHILDREN = 25

MUTATION_RATE = 0.15

MUTATION_BEST_RATE = 0.1

MUTATION_CHANGE_OVER_EPOCHS = 120

MUTATION_DECREASE = 0.015

INCREASE_IF_NO_IMPROVES = 80

MUTATION_INCREASE = 0.0075

CHILDREN_INCREASE = 2
```

#### Now we are going to run our genetic algorithm. On every epoch we create N children from individual and select only 1 best for next epochs.

```python
ind = create_individual(df)

stable_score = 0

best_score = fitness(MAX_MONEY, AVG_AGE, ind)

best_ind = ind

for i in range(EPOCHS):
    
    improve_flag = False
    
    if stable_score == INCREASE_IF_NO_IMPROVES:
        
        stable_score = 0
        
        print('Mutation rate increased to: ', MUTATION_RATE + MUTATION_INCREASE)
        
        MUTATION_RATE += MUTATION_INCREASE
        
        CHILDREN += CHILDREN_INCREASE
        
    if i % 20 == 0:
        
        print('Epoch: ', i, best_score)
        
    if i % MUTATION_CHANGE_OVER_EPOCHS == 0 and i != 0:
        
        print('Mutation rate decreased to: ', MUTATION_RATE - MUTATION_DECREASE)
        
        MUTATION_RATE -= MUTATION_DECREASE
        
    for i in range(CHILDREN):
        
        child = mutate(ind, MUTATION_RATE, MUTATION_BEST_RATE, df)
        
        child_score = fitness(MAX_MONEY, AVG_AGE, child)
        
        if child_score >= best_score:
            
            best_score = child_score
            
            best_ind = child
            
            improve_flag = True
            
    if improve_flag == False:
        
        stable_score += 1
```

#### Finally - it is our best team.

```python
for player in best_ind:
    
    print(player.name, player.overall)
```

#### Storing First and Second Team in order

```python
first_team = []

second_team = []

i = 0

while i < len(best_ind):
    
    if best_ind[i].overall >= best_ind[i+1].overall:
        
        first_team.append(best_ind[i])
        
        second_team.append(best_ind[i+1])
        
    else:
        
        first_team.append(best_ind[i+1])
        
        second_team.append(best_ind[i])
        
    i+=2
    
first_team.reverse()

second_team.reverse()
```
